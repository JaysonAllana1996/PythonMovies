```python
# First let's import packages that we will need for the project
import pandas as pd
import numpy as np
import seaborn as sns

```

```python
# Import dataset into a dataframe
df=pd.read_csv(r'C:\Users\Jayson Naquila\Desktop\Dataset files\movies 1986-2016\movies.csv')
df.head()

```

```python
# Finding out null values per column
for col in df.columns:
    print(df[col].notnull().value_counts(), "\n")
    
```

In the budget columns, there are 2171 null values vs 5497 not null values. Dropping 2171 rows is a lot of data. We will keep the rows with null values for now.

```python
# Getting the percentage of null values by column
for col in df.columns:
    null_pct=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,null_pct*100))
```

The budget column has a total of 28% null values. That is a lot of data if dropped.


```python
# We have to drop all rows with NA because python cannot process null data
df=df.dropna()
for col in df.columns:
    print(df[col].notnull().value_counts(), "\n")
```

```python
# Checking percentage of null data after dropping all nulls
for col in df.columns:
    null_pct=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,null_pct*100))
```

We now have a clean data frame without nulls.

```python
# Let's check data types of each columms
df.dtypes

```

```python
df.head()
```

Votes, budget, gross, and runtime is float type. Let's change the data type of those columns into integer.

```python
# Converting float types into integers
df['votes']=df['votes'].astype('int64')
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df['runtime']=df['runtime'].astype('int64')
df.head()
```

```python
df.dtypes
```

Now all float types are converted into integers.

```python
# Importing libraries of graphing
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Exploring total numbers of movies made by directors
directors_movie=df.groupby('director')['name'].count()
directors_movie=directors_movie.sort_values(ascending=False)
print(directors_movie)
```

Woody Allen tops the director with the most number of movies made.

```python
# Most profitable directors ranking
directors_rank=df.groupby('director')['gross'].sum()
directors_rank=directors_rank.sort_values(ascending=False)
print(directors_rank)
```

Although Woody Allen made the most films, Steven Spielberg still tops the gross ranking.

```python
# Graphing directors with their total number of movies

directors_movie=df.groupby('director')['name'].count()
plt.figure(figsize=(10, 6))
plt.hist(directors_movie, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Number of Movies')
plt.ylabel('Number of Directors')
plt.title('Distribution of Movies per Director')
plt.grid(axis='y')
plt.show()
```

More than 1400 directors have made only one film, and only two directors have made 30 and above films.

Exploring Data Correlation 

```python

df.corr(numeric_only=True, method='pearson')
```

```python
corr_matrix=df.corr(numeric_only=True, method='pearson')
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt='.2f')
```

Let's explore correlations between values in the dataset. The closer the correlation to 1, the more related the values are.

```python
#Plotting budget vs gross correlation

plt.scatter(df['budget'], df['gross'])
plt.xlabel('budget')
plt.ylabel('gross')
plt.title('Scatter Plot')
plt.show()
```

```python
#Plotting budget vs gross regression
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})

plt.xlabel('Budget')
plt.ylabel('Gross')
plt.title('Plotting budget vs gross regression')
plt.show()
```

```python
#Plotting score vs gross regression
sns.regplot(x='score', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})

plt.xlabel('score')
plt.ylabel('Gross')
plt.title('Scatter Plot score vs gross regression')
plt.show()
```

```python
#Plotting score vs total votes regression
sns.regplot(x='score', y='votes', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})

plt.xlabel('score')
plt.ylabel('votes')
plt.title('Plotting score vs total votes regression')
plt.show()
```

```python
#Plotting votes vs total gross regression
sns.regplot(x='votes', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})

plt.xlabel('votes')
plt.ylabel('gross')
plt.title('Plotting votes vs total gross regression')
plt.show()
```

```python
#Plotting runtime vs score regression
sns.regplot(x='runtime', y='score', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})

plt.xlabel('runtime')
plt.ylabel('score')
plt.title('Scatter Plot  runtime vs score with Regression Line')
plt.show()
```
